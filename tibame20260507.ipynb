```python
row = 4
col = 5

# j=0~4
# i=0 [0, 1, 2, 3, 4]
# i=1 [9, 8, 7, 6, 5]
# i=2 [10, 11, 12, 13, 14]
# i=3 [19, 18, 17, 16, 15]
# i=0 0+j  i=2 10+j i=4 20+j
result = [[0] * col for i in range(row)]

start = 3
for i in range(row):
    for j in range(col):
        if i % 2 == 0:
            result[i][j] = col * i + j + start
        else:
            result[i][j] = (col * i + (col - 1)) - j + start

result
```

1. 群集型態: list/dict/set/tuple

2. list: 同類型的清單 [學生, 學生, 學生], list的key是固定的 0 1 2 3 (有序)

3. dict: 聚集多個欄位(key:value)來成為複雜單位 {"name":"Elwing", "身高":175} (無序)

4. set: 同類型但不重複的集合(無序) {30, 20, 15, 30} 沒有key

5. tuple: 簡化版本的字典 {"r":255, "g":128, "b":200}(255, 128, 200) key=0 1 2  操作: tuple[key]

```python
n = {30, 20, 15, 30}
print(n)
# 交集(&): A B 兩集合都有的
# 聯集(|): 把 A B 有的統整起來
# 減集(-): A - B 把 A 的東西扣掉 B 有的(A - AB交集)
a = {20, 30, 40}
b = {40, 50, 60}
print(a & b)
print(a | b)
print(a - b)
# 專屬功能: add
a.add(90)
print(a)
# 迴圈
for n in (a | b):
    print("!", n)
```

```python
import random
# 如果用list
numbers = []
while True:
    n = random.randint(1, 49)
    # print(n)
    if not n in numbers:
        numbers.append(n)
    if len(numbers) == 6:
        break
print(numbers)
# 如果你用set
numbers = set()
while True:
    n = random.randint(1, 49)
    numbers.add(n)
    if len(numbers) == 6:
        break
print(numbers)
```
