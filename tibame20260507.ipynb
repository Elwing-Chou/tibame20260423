```python
# 雙層清單
result = [[0] * 5 for i in range(3)]
print(result)
# 雙層清單
result = []
for i in range(3):
    result.append([0] * 5)
print(result)
```

```python
row = 4
col = 5

# j=0~4
# i=0 [0, 1, 2, 3, 4]
# i=1 [9, 8, 7, 6, 5]
# i=2 [10, 11, 12, 13, 14]
# i=3 [19, 18, 17, 16, 15]
# i=0 0+j  i=2 10+j i=4 20+j
result = [[0] * col for i in range(row)]

start = 3
for i in range(row):
    for j in range(col):
        if i % 2 == 0:
            result[i][j] = col * i + j + start
        else:
            result[i][j] = (col * i + (col - 1)) - j + start

result
```
