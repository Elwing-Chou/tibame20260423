```python
row = 4
col = 5

# j=0~4
# i=0 [0, 1, 2, 3, 4]
# i=1 [9, 8, 7, 6, 5]
# i=2 [10, 11, 12, 13, 14]
# i=3 [19, 18, 17, 16, 15]
# i=0 0+j  i=2 10+j i=4 20+j
result = [[0] * col for i in range(row)]

start = 3
for i in range(row):
    for j in range(col):
        if i % 2 == 0:
            result[i][j] = col * i + j + start
        else:
            result[i][j] = (col * i + (col - 1)) - j + start

result
```

1. 群集型態: list/dict/set/tuple

2. list: 同類型的清單 [學生, 學生, 學生], list的key是固定的 0 1 2 3 (有序)

3. dict: 聚集多個欄位(key:value)來成為複雜單位 {"name":"Elwing", "身高":175} (無序)

4. set: 同類型但不重複的集合(無序) {30, 20, 15, 30} 沒有key

5. tuple: 簡化版本的字典 {"r":255, "g":128, "b":200}(255, 128, 200) key=0 1 2  操作: tuple[key]

```python
n = {30, 20, 15, 30}
print(n)
# 交集(&): A B 兩集合都有的
# 聯集(|): 把 A B 有的統整起來
# 減集(-): A - B 把 A 的東西扣掉 B 有的(A - AB交集)
a = {20, 30, 40}
b = {40, 50, 60}
print(a & b)
print(a | b)
print(a - b)
# 專屬功能: add
a.add(90)
print(a)
# 迴圈
for n in (a | b):
    print("!", n)
```

```python
import random
# 如果用list
numbers = []
while True:
    n = random.randint(1, 49)
    # print(n)
    if not n in numbers:
        numbers.append(n)
    if len(numbers) == 6:
        break
print(numbers)
# 如果你用set
numbers = set()
while True:
    n = random.randint(1, 49)
    numbers.add(n)
    if len(numbers) == 6:
        break
print(numbers)
```

```python
import pygame as pg
import random

#pygame初始化
pg.init()

# --- 字型 ---
def get_font(size):
    for f in ['microsoftjhenghei', 'simhei', 'stheitirelight']:
        if f in pg.font.get_fonts():
            return pg.font.SysFont(f, size)
    return pg.font.SysFont(None, size)


# GAME UI
FONT_UI = get_font(32)
col, row = 15, 15
inter = 40
width, height = (col + 2) * inter, (row + 2) * inter
# 產生視窗
screen = pg.display.set_mode([width, height])
# 設定遊戲標題
pg.display.set_caption("踩地雷")

# 遊戲的邏輯
BOMB_COUNT = 20
NOT_CLICKED = -1
# 產生20個不重複的位置當作我的地雷
BOMB = -99
game_board = [[NOT_CLICKED] * col for i in range(row)]
bombs = set()
while True:
    i = random.randint(0, row-1)
    j = random.randint(0, col-1)
    bombs.add((i, j))
    game_board[i][j] = BOMB
    if len(bombs) == BOMB_COUNT:
        break

# 準備第一個圖層
bg = pg.Surface(screen.get_size())
# 把畫布填滿某個顏色
bg.fill([199, 167, 82])


# 2. 產生並且畫雷
# pygame.draw.rect(畫布, 顏色, [x坐標, y坐標, 寬度, 高度], 線寬)
pg.draw.rect(bg, [255, 0, 0], [80, 200, inter, inter], 0)

# 3. 當點擊的時候畫出數字
pg.draw.rect(bg, [255, 255, 255], [120, 200, inter, inter], 0)
# 字體(字, 平滑話, 顏色)
t = FONT_UI.render("4", 1, [0, 0, 0])
# 背景(bg)上面疊上t, 字體(center) -> get_rect -> (左上x, 左上y, 寬, 高)
bg.blit(t, t.get_rect(center=[140, 220]))

# 1. 畫棋盤的線
# pygame.draw.line(畫布, 顏色, (x坐標1, y坐標1), (x坐標2, y坐標2), 線寬)
for i in range(row+1):
    pg.draw.line(bg,
                 [0, 0, 0],
                 [inter, inter*i+inter],
                 [width-inter, inter*i+inter],
                 1)

for i in range(col+1):
    pg.draw.line(bg,
                 [0, 0, 0],
                 [inter*i+inter, inter],
                 [inter*i+inter, height-inter],
                 1)

# 你要把圖層放到上一層
screen.blit(bg, [0, 0])
# 對畫面進行更新(才會真的秀出來)
pg.display.update()


# 建立一個永不結束的迴圈(遊戲才不會結束)
running = True
while running:
    # 收取你的遊戲任何事件(滑鼠點擊/鍵盤按鈕...)
    for event in pg.event.get():
        # 偵測滑鼠點擊以後放掉的動作
        if event.type == pg.MOUSEBUTTONUP:
            # 1. 需要一個儲存媒介, 告訴我點下去的位置到底是安全還是地雷
            # 2. 需要一個近似方式, 來把我點的位置近似到某一個框框
            # (看我點的位置跟哪個框的中心點最近): 為了方便, 我要順便把每一個框的中心點位置儲存起來
            pass
        # 如果收到的事件是按x
        if event.type == pg.QUIT:
            # 迴圈就會變成while False
            running = False

pg.quit()
```
