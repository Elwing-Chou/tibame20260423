```python
def is_leap_year(year):
    # year = 2000
    if year % 400 == 0:
        return True
    elif year % 100 == 0:
        return False
    elif year % 4 == 0:
        return True
    else:
        return False
```

```python
def is_date_valid(d):
    day = d % 100
    remain = d // 100
    month = remain % 100
    year = remain // 100

    if year >= 1:
        if 12 >= month >= 1:
            month_days = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
            days_max = month_days[month-1]
            # leap year: if is_learp_year(year) == True
            if is_leap_year(year) and month == 2:
                days_max = days_max + 1
            if days_max >= day >= 1:
                return True
            else:
                return False
        else:
            return False
    else:
        return False

print(is_date_valid(20000229))
print(is_date_valid(20010229))
```

1. 走過群集: for 名 in 群

3. 不固定次數: while True+break

```python
start = 4499
result = []
count = 0
while True:
    d = start * start
    if is_date_valid(d):
        # 清單的專屬功能: append, 直接修改
        # 錯誤寫法: result = result.append
        result.append(d)
        count = count + 1
    if count == 10:
        break
    start = start + 1
print(result)
```

```python
def get_squared_2(start=4499, max_count=10):
    result = []
    count = 0
    while True:
        d = start * start
        if is_date_valid(d):
            # 清單的專屬功能: append, 直接修改
            # 錯誤寫法: result = result.append
            result.append(d)
            count = count + 1
        if count == max_count:
            break
        start = start + 1
    return result

print(get_squared_2(max_count=20))
```

1. python處理時間的內鍵模組: datetime, time

2. datetime: a. str<->datetime b. 日期的增加減少

```python
import datetime as dt
# 型態轉換: str <-> datetime
# strptime: str->datetime
# strftime: datetime->str
s = "2000-02-29"
d = dt.datetime.strptime(s, "%Y-%m-%d")
s2 = d.strftime("會議時間: %Y/%m/%d(%a)")
print(s2)
```

```python
# 日期的增加減少
d = dt.datetime(2000, 2, 28)
delta = dt.timedelta(days=1)
d + 10 * delta
```
