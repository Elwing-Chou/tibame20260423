```python
def is_leap_year(year):
    # year = 2000
    if year % 400 == 0:
        return True
    elif year % 100 == 0:
        return False
    elif year % 4 == 0:
        return True
    else:
        return False
```

```python
def is_date_valid(d):
    day = d % 100
    remain = d // 100
    month = remain % 100
    year = remain // 100

    if year >= 1:
        if 12 >= month >= 1:
            month_days = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
            days_max = month_days[month-1]
            # leap year: if is_learp_year(year) == True
            if is_leap_year(year) and month == 2:
                days_max = days_max + 1
            if days_max >= day >= 1:
                return True
            else:
                return False
        else:
            return False
    else:
        return False

print(is_date_valid(20000229))
print(is_date_valid(20010229))
```

```python
start = 4499
result = []
count = 0
while True:
    d = start * start
    if is_date_valid(d):
        # 清單的專屬功能: append, 直接修改
        # 錯誤寫法: result = result.append
        result.append(d)
        count = count + 1
    if count == 10:
        break
    start = start + 1
print(result)
```

```python
def get_squared_2(start=4499, max_count=10):
    result = []
    while True:
        d = start * start
        if is_date_valid(d):
            # 清單的專屬功能: append, 直接修改
            # 錯誤寫法: result = result.append
            result.append(d)
        if len(result) == max_count:
            break
        start = start + 1
    return result

print(get_squared_2(max_count=20))
```

```python
import datetime as dt
start = dt.datetime(2024, 10, 1)
delta = dt.timedelta(days=1)
# 一樣放入20241001 2072..整數
result = []
while True:
    # 日期->字串->整數
    # 開根號->確認is_integer
    # 數目+1
    d = int(start.strftime("%Y%m%d"))
    d_root = d ** 0.5
    if d_root.is_integer():
        result.append(d)
    if len(result) == 10:
        break
    start = start + delta
print(result)
```

```python
# 把想法1 變成函式
def get_squared_1(y=2024, m=10, d=1, max_count=10):
    start = dt.datetime(y, m, d)
    delta = dt.timedelta(days=1)
    # 一樣放入20241001 2072..整數
    result = []
    while True:
        # 日期->字串->整數
        # 開根號->確認is_integer
        # 數目+1
        d = int(start.strftime("%Y%m%d"))
        d_root = d ** 0.5
        if d_root.is_integer():
            result.append(d)
        if len(result) == max_count:
            break
        start = start + delta
    return result

get_squared_1(max_count=5)
```

```python
import time

count = 300

t1 = time.time()
get_squared_1(max_count=count)
t2 = time.time()
t_dif_1 = t2 - t1
print("想法1:", t_dif_1, "秒")

t1 = time.time()
get_squared_2(max_count=count)
t2 = time.time()
t_dif_2 = t2 - t1
print("想法2:", t_dif_2, "秒")

ratio = t_dif_1 / t_dif_2
print("相差幾倍:", ratio)
```
